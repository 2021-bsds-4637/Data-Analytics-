```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset

df = pd.read_csv(r"C:\Users\Hp\Downloads\Accommodation.csv")
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
# Specify the columns to drop null values
columns_to_drop_null = ['Url', 'Telephone', 'AddressLocality']

# Drop null values from the specified columns
df.dropna(subset=columns_to_drop_null, inplace=True)
```

```python
# Get the counts of unique values for both 'AddressRegion' and 'AddressLocality'
region_counts = df['AddressRegion'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(6, 2))

region_counts.plot(kind='bar', color='lightgrey')
plt.title('Counts of Address Regions')
plt.ylabel('Count')
plt.xlabel('Address Region')
plt.grid(False)  

plt.show()
```

```python
plt.figure(figsize=(4, 7))
sns.scatterplot(data=df, x='Longitude', y='Latitude', hue='AddressCountry', palette='viridis')
plt.title('Accommodation Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Country')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Tags', order=df['Tags'].value_counts().index, palette='Set2')
plt.title('Tag Distribution')
plt.xlabel('Count')
plt.ylabel('Tags')
plt.show()
```
